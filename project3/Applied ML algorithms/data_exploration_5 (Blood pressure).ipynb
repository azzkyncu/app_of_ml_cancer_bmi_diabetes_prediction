```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import tree
```

```python
#renaming the columns 
data.columns = ['state', 'state_code', 'sex', 'marital_status', 'age', 'race',
       'education', 'weight', 'height', 'income', 'vegetables_day',
       'fruits_day', 'physical_ctivity', 'smoking', 'alcohol_day',
       'bmi', 'blood_pressure', 'cholestrol', 'heart_disease', 'stroke',
       'cancer', 'diabetes']
```

```python
data.head()
```

```python
data.dropna(axis='columns')
```

```python
# dropping 'refused' and ''DonÂ´t Know/Not Sure' from all rows. 
for col in data.columns:
    data = data[data[col]!="Refused"]
    data = data[data[col]!="Don't Know/Not Sure"]
    data = data[data[col]!="Donâ€™t know/Not sure"]
    data = data[data[col]!="DonÂ´t know/Not Sure"]
    data = data[data[col]!="Donâ€™t know/Not sure/Refused"]
    data = data[data[col]!="Donâ€™t know/Not Sure"]
```

```python
data.head()
```

```python
data['blood_pressure'].value_counts()
```

```python

Y = data['blood_pressure'].copy()
Y.value_counts(normalize=True)
Y = pd.get_dummies(Y,drop_first=False)
Y = Y.rename(columns={"Yes":"blood_pressure"})
del Y["No"]
del Y["Borderline"]
Y.head()
```

```python
Y['blood_pressure'].value_counts(normalize=True)
```

```python
X = data.copy()
del X['blood_pressure']
del X['state']
del X['state_code']
```

```python
# Ordinal encode age
age_map = {
    '18-24':0,
    '25-29':1,
    '30-34':2,
    '35-39':3,
    '40-44':4,
    '45-49':5,
    '50-54':6,
    '55-59':7,
    '60-64':8,
    '65-69':9,
    '70-74':10,
    '75-79':11,
    '>80':12}

X = X.replace({'age':age_map})
X['age'] = X['age'].astype(int)
```

```python
# Ordinal encode age
inc_map = {
    '< 10000':0,
    '10000-15000':1,
    '15000-20000':2,
    '20000-25000':2,
    '25000-35000':3,
    '35000-50000':4,
    '50000-75000':5,
    ' >75000':6}
X = X.replace({'income':inc_map})
X['income'] = X['income'].astype(int)
X
```

```python
X['income'].value_counts()
```

```python
X = pd.get_dummies(X, drop_first=True)
X.columns
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state=20)
X_res, y_res = sm.fit_resample(X_train, y_train.values.ravel())
print(X_res.shape)
print(y_res.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_res, y_res)
rf.score(X_test, y_test)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estimators=50)
xgb.fit(X_res, y_res)
print(xgb)
```

```python
print(xgb.score(X_test.to_numpy(), y_test.to_numpy()))
```

```python
y_pred = xgb.predict(X_test.to_numpy())
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_true = y_test, y_pred = y_pred)
```
